```python
from bs4 import BeautifulSoup as bsoup
import requests as rq
import re
```

```python
def get_url_list():
    
    url_list = []

    for i in range(1, 24):
        base_url = 'https://editions.flammarion.com/Catalogue/(collection)/GF--4315/(sous-domaine)/Litt%C3%A9rature%20classique/(serie)/Litt%C3%A9rature%20et%20civilisation--3383/(page)/'+str(i)+'/(tri)/parution_asc'
        url_list.append(base_url)
    return url_list
```

```python
url_list = get_url_list()
```

```python
def get_titles(url_list):
    results = []
    for url_ in url_list:
        r_new = rq.get(url_)
        soup_new = bsoup(r_new.text)
        arbo_titres = soup_new.find_all(class_="BookList-title")
        for title in arbo_titres:
            results.append([a.text for a in title.find_all("a")])
    return results
```

```python
list_titles = get_titles(url_list)
```

```python
def convert_titles(list_to_convert):
    liste_titres = []
    for title in list_to_convert:
        liste_titres.append(title[0])
    return liste_titres
```

```python
liste_titres = convert_titles(list_titles)
```
