```python
df_auteurs = pd.read_csv(r'data/features_canon.csv')
```

```python
df_auteurs_canon = df_auteurs[df_auteurs['canon']=='canon']
```

```python
def get_canon(df): 
    list_auteur = []
    list_date = []
    list_titre = []
    
    for doc_name in df['index']:
        date = doc_name.split("_")[0]
        list_date.append(date)
        auteur = doc_name.split("_")[1]
        list_auteur.append(auteur)
        titre = doc_name.split("_")[2]
        list_titre.append(titre)
        
    zipped = list(zip(list_date, list_auteur, list_titre)) 
    df = pd.DataFrame(zipped, columns=['Date', 'Auteur', 'Titre'])
    
    return df
```

```python
df_canonical = get_canon(df_auteurs_canon)
```

```python
df_canonical
```

```python
sorted_df = df_canonical.sort_values(by='Date')
```
